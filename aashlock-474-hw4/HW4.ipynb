```python
from sklearn.cluster import KMeans
import numpy as np

X = np.array([[0,0],[10,10],[100,100],[1,1],[0,5],[-3,4],[-1,1],[0,-10],[30,-4]])
kmeans = KMeans(n_clusters=2, n_init="auto").fit(X)
clusters = kmeans.fit_predict(X)

cluster0 = X[clusters==0]
cluster1 = X[clusters==1]
 
plt.scatter(cluster0[:,0] , cluster0[:,1] , color = 'red')
plt.scatter(cluster1[:,0] , cluster1[:,1] , color = 'blue')
plt.show()
```

```python
import numpy as np

X = np.array([[3,5],[-2,10],[9,-5],[1,7],[0,-2],[-3,4],[-1,3],[0,0],[2,-4]])
Y = np.array([-1,1,1,1,-1,1,-1,1,1])
label1 = X[Y==1]
labelneg1 = X[Y==-1]

plt.scatter(label1[:,0], label1[:,1], color = 'red')
plt.scatter(labelneg1[:,0], labelneg1[:,1], color = 'blue')

plt.legend(["Y = 1", "Y = -1"], loc="upper right")
plt.show()
```

```python
import numpy as np
from sklearn.svm import SVC
from sklearn.inspection import DecisionBoundaryDisplay

X = np.array([[3,5],[-2,10],[9,-5],[1,7],[0,-2],[-3,4],[-1,3],[0,0],[2,-4]])
y = np.array([-1,1,1,1,-1,1,-1,-1,1])        
label1 = X[y==1]
labelneg1 = X[y==-1]

classifier = SVC(kernel='linear').fit(X, y)
#print(classifier.predict([[2,0]]))

plt.scatter(label1[:,0], label1[:,1], color = 'red')
plt.scatter(labelneg1[:,0], labelneg1[:,1], color = 'blue')

# Step 4: Plot decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
Z = classifier.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.8, colors=['lightblue', 'pink'])

plt.legend(["Y = 1", "Y = -1"], loc="upper right")
plt.show()
```
